## Movimento Browniano Geométrico na Previsão do valor de Ações

### Import de biblotecas necessárias

```python
import matplotlib.pyplot as plt
import yfinance as yf
import numpy as np
import pandas as pd
import datetime as dt
from scipy.stats import lognorm, anderson
```

### Aplicação do modelo na ação da Amazon com dados de 01/05/2021 a 01/05/2022. Previsão a um ano.

```python
plt.style.use("fivethirtyeight")
ticker = "AMZN"
start = dt.datetime(2021, 5, 1)
end = dt.datetime(2022,5,1)
stock_data = yf.download(ticker, start, end)
returns = stock_data['Adj Close'].pct_change()

start_test = dt.datetime(2022, 5, 1)
end_test = dt.datetime(2023,5,1)
stock_data_test = yf.download(ticker, start_test, end_test)

std = returns.std()
mu = returns.mean() 
t = 1

fig,ax = plt.subplots(figsize = (13,7))


big_list = []

for i in range(5000):
    stock_value = [stock_data["Adj Close"][-1]]
    for day in range(len(stock_data_test["Adj Close"])-1):
        value = stock_value[-1] * np.exp((mu - (std**2)/2) * t + std * np.random.normal(0,1))
        stock_value.append(value)
    ax.plot(range(len(stock_data_test["Adj Close"])), stock_value, alpha=0.13, linewidth=0.9, zorder = 1, color="#ADADAD")
    big_list.append(stock_value)

ax.plot(range(len(stock_data_test["Adj Close"])), stock_value, alpha=0.13, linewidth=0.9, zorder = 1, color="#ADADAD", label = "Simulações")


df = pd.DataFrame(big_list, columns=range(len(stock_data_test["Adj Close"])))
df.loc["mean"] = df.median()
df.loc["std"] = df.std()


ax.plot(range(len(stock_data_test["Adj Close"])), df.loc["mean"], linewidth=2.5, c ="k", label="Média das Simulações")
ax.plot(range(len(stock_data_test["Adj Close"])), stock_data_test["Adj Close"], linewidth = 3, c = "r", label = "Valor Real")
ax.plot(range(len(stock_data_test["Adj Close"])), df.loc["mean"] - df.loc["std"]*1.96/2, color = "k", linestyle = "dotted", linewidth = 2, label="Intervalo de Confiança de 95%")
ax.plot(range(len(stock_data_test["Adj Close"])), df.loc["mean"] + df.loc["std"]*1.96/2, color = "k", linestyle = "dotted", linewidth = 2)
#ax.fill_between(range(366), df.loc["mean"] - df.loc["std"]*1.96/2, df.loc["mean"] + df.loc["std"]*1.96/2, color = "#26A1EE", alpha = 0.2, zorder = 2)

ax.set_title("Previsão de valor de ação da Amazon", loc = "left")
ax.set_xlabel("Dias de trading após 1 de Maio de 2022")
ax.set_ylabel("Valor em Doláres Americanos ($)")

plt.legend(fancybox=False)
plt.show()
```

### Histograma da distribuição dos valores obtidos após um ano de caminhada aleatória. Ajuste lognormal.

```python
fig,ax = plt.subplots(figsize = (13,7))
ax.hist(df.iloc[:,-1], bins = 50, color = "#ADADAD", density = True, label = "Distribuição de valores")
ax.set_title("Histograma dos valores previstos para o valor da ação da Amazon", loc = "left")
ax.set_xlabel("Valor em Doláres Americanos ($)")

expected = np.mean(df.iloc[:,-1])
x = np.linspace(np.min(df.iloc[:,-1]), np.max(df.iloc[:,-1]), 100)
shape,loc,scale = lognorm.fit(df.iloc[:,-1])
pdf = lognorm.pdf(x, shape, loc, scale)
ax.plot(x,pdf, color = "red", label = "Ajuste lognormal")
ax.axvline(expected, label = "Valor Esperado")

plt.legend()
plt.show()
```

### Teste de erro médio percentual e percentagem de valores no intervalo de confiança de 95% para ações do S&P500

```python
from pytickersymbols import PyTickerSymbols
tickers = []

for index in ["S&P 500"]:
    stock_data = PyTickerSymbols()
    stocks = stock_data.get_stocks_by_index(index)
    df = pd.DataFrame.from_dict(list(stocks), orient='columns')
    tickers += list(df["symbol"])
```

```python
df_results = []
N = 1500

for ticker in tickers:
    try:
        stock_data = yf.download(ticker, start, end)   
        stock_value = [stock_data["Adj Close"][-1]]
    except:
        continue
    start = dt.datetime(2015, 5, 1)
    end = dt.datetime(2016,5,1)
    stock_data = yf.download(ticker, start, end)   
    returns = stock_data['Adj Close'].pct_change()

    start_test = dt.datetime(2016,5,1)
    end_test = dt.datetime(2017,5,1)
    stock_data_test = yf.download(ticker, start_test, end_test)

    std = returns.std()
    mu = returns.mean() 
    

    t = 1

    big_list = []
    
    for i in range(N):
        stock_value = [stock_data["Adj Close"][-1]]
        for day in range(len(stock_data_test["Adj Close"])-1):
            value = stock_value[-1] * np.exp((mu - (std**2)/2) * t + std * np.random.normal(0,1))
            stock_value.append(value)
        big_list.append(stock_value)
    try:
        df = pd.DataFrame(big_list, columns=range(len(stock_data_test["Adj Close"])))
    except:
        continue
    df.loc["mean"] = df.median()
    df.loc["std"] = df.std()

    mpe = np.sum((df.iloc[:N,-1] - stock_data_test["Adj Close"][-1]) / stock_data_test["Adj Close"][-1] * 100) / N
    in_95ic = stock_data_test["Adj Close"][-1]>df.loc["mean"].iloc[-1]-df.loc["std"].iloc[-1]*1.96/2 and stock_data_test["Adj Close"][-1]<df.loc["mean"].iloc[-1]+df.loc["std"].iloc[-1]*1.96/2
    df_results.append([mpe, in_95ic, std])
```

```python
df_results = pd.DataFrame(df_results, columns=["MPE", "Is in 95% CI", "std"])
df_results.loc["Results"] = [np.mean(np.abs(df_results["MPE"])), sum(df_results["Is in 95% CI"])/len(df_results["Is in 95% CI"]), np.mean(df_results["std"])]
df_results = df_results[df_results["MPE"] < 1000]
df_results = df_results[df_results["MPE"] > -99]
df_results = df_results[df_results["std"] < 50]
df_results.loc["Results"] = [np.mean(np.abs(df_results["MPE"])), sum(df_results["Is in 95% CI"])/len(df_results["Is in 95% CI"]), np.mean(df_results["std"])]
```

```python
quantile_25 = np.quantile(df_results["std"], 0.25)
quantile_75 = np.quantile(df_results["std"], 0.75)

df_lower_std = df_results[df_results["std"] < quantile_25]
df_higher_std = df_results[df_results["std"] > quantile_75]
```

```python
df_lower_std.loc["Results"] = [np.mean(np.abs(df_lower_std["MPE"])), sum(df_lower_std["Is in 95% CI"])/len(df_lower_std["Is in 95% CI"]), np.mean(df_lower_std["std"])]
df_higher_std.loc["Results"] = [np.mean(np.abs(df_higher_std["MPE"])), sum(df_higher_std["Is in 95% CI"])/len(df_higher_std["Is in 95% CI"]), np.mean(df_higher_std["std"])]
```
